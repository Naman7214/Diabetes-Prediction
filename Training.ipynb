```python
X = df_final.drop('Diabetes_binary',axis = 1)
X = X.drop('index',axis = 1)
X = X.drop('Education', axis = 1)
X = X.drop('Income', axis =1 )
Y = df_final['Diabetes_binary']
```

```python
(df_final['Diabetes_binary']).unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 23)
```

```python
###USING CLASICAL MACHINE LEARNING MODEL
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC = RandomForestClassifier()
```

```python
RFC.fit(X_train,Y_train)
```

```python
Y_pred = RFC.predict(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report,accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(Y_test,Y_pred)
```

```python
print(classification_report(Y_test,Y_pred))
print(accuracy_score(Y_test,Y_pred))
```

```python
### USING NEURAL NETWORK
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=256, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs=100, batch_size=512)
```

```python
import matplotlib.pyplot as plt
loss = history.history['loss']
plt.plot(loss, label='Training Loss')
plt.title('Training Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
Ypred = model.predict(X_test)
```

```python
Ypred
```

```python
Ypred= (Ypred >= 0.5).astype(int)
```

```python
print(classification_report(Y_test,Ypred))
print(accuracy_score(Y_test,Ypred))
```

```python
ConfusionMatrixDisplay.from_predictions(Y_test,Ypred)
plt.show()
```
